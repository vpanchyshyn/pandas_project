```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Odata2021File.csv', sep=';')
dataset
```

```python
"Calculate the total number of participants"
participants_count = len(dataset.index)

f'Загальна кількість учасників ЗНО 2021: {participants_count}'
```

```python
"Show histogram of birth year of participants and calculate the min and max age."
age_max = min(dataset.iloc[:, 1])
age_min = max(dataset.iloc[:, 1])

print(f'Максимальний вік учасника: {2021-age_max} ({age_max} р.н.)')
print(f'Мінімальний вік учасника: {2021-age_min} ({age_min} р.н.)')
dataset['Birth'].plot(kind='hist')
plt.title('Histogram: birth year of participants')
plt.show()
```

```python
"Show pie charts with percentage of students that passed and failed main exams."
passed_ukr = dataset['UMLTestStatus'].value_counts()['Зараховано']
failed_ukr = dataset['UMLTestStatus'].value_counts()['Не подолав поріг']

passed_math = dataset['MathTestStatus'].value_counts()['Зараховано']
failed_math = dataset['MathTestStatus'].value_counts()['Не подолав поріг']

passed_hist = dataset['HistTestStatus'].value_counts()['Зараховано']
failed_hist = dataset['HistTestStatus'].value_counts()['Не подолав поріг']

passed_eng = dataset['EngTestStatus'].value_counts()['Зараховано']
failed_eng = dataset['EngTestStatus'].value_counts()['Не подолав поріг']

labels = ['Склав', 'Не склав']
sizes_ukr = [passed_ukr, failed_ukr]
sizes_math = [passed_math, failed_math]
sizes_hist = [passed_hist, failed_hist]
sizes_eng = [passed_eng, failed_eng]

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].pie(sizes_ukr, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0, 0].set_title('Результати: Українська мова та література')

axs[0, 1].pie(sizes_math, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0, 1].set_title('Результати: Математика')

axs[1, 0].pie(sizes_hist, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 0].set_title('Результати: Історія')

axs[1, 1].pie(sizes_eng, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Результати: Англійська мова')
plt.tight_layout()
plt.show()

```

```python
"Show the histogram of number of students for each subject."
ukr = dataset['UkrTest'].value_counts()['Українська мова']
uml = dataset['UMLTest'].value_counts()['Українська мова і література']
hist = dataset['HistTest'].value_counts()['Історія України']
math = dataset['MathTest'].value_counts()['Математика']
phys = dataset['PhysTest'].value_counts()['Фізика']
chem = dataset['ChemTest'].value_counts()['Хімія']
bio = dataset['BioTest'].value_counts()['Біологія']
geo = dataset['GeoTest'].value_counts()['Географія']
eng = dataset['EngTest'].value_counts()['Англійська мова']
fr = dataset['FraTest'].value_counts()['Французька мова']
deu = dataset['DeuTest'].value_counts()['Німецька мова']
sp = dataset['SpaTest'].value_counts()['Іспанська мова']

subjects = ['Ukr', 'UML', 'Hist', 'Math', 'Phys', 'Chem', 'Bio', 'Geo', \
'Eng', 'Fr', 'Deu', 'Sp']
values = [ukr, uml, hist, math, phys, chem, bio, geo, eng, fr, deu, sp]
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink', \
'gray', 'cyan', 'magenta', 'yellow', 'lightgreen']

plt.bar(subjects, values, color=colors)
plt.rcParams.update({'font.size': 12}) 
plt.xlabel('Предмети')
plt.ylabel('Кількість учнів')
plt.title('Кількість учнів, що складали кожен предмет')

plt.show()

```

```python
"""
Show the histogram of maximum and minimum average results in 4 subjects
in regions.
"""
names = ['Українська мова і література', 'Математика', 'Англійська мова', 'Історія України']
subjects = ['UMLBall100', 'MathBall100', 'EngBall100', 'HistBall100']
result = {'Subject': [], 'Max': [], 'Min': [], 'MaxRegion': [], 'MinRegion': []}

for subj in subjects:
    numeric_series = pd.to_numeric(dataset[subj].str.replace(',', '.', regex=False), errors='coerce')
    numeric_df = pd.DataFrame({'Score': numeric_series, 'RegName': dataset['RegName']})
    numeric_df = numeric_df[(numeric_df['Score'].notna()) & (numeric_df['Score'] != 0)]
    result_df = numeric_df.groupby('RegName')['Score'].mean().reset_index()
    max_mean_region = result_df.loc[result_df['Score'].idxmax()]
    min_mean_region = result_df.loc[result_df['Score'].idxmin()]
    result['Subject'].append(subj)
    result['Max'].append(max_mean_region['Score'])
    result['Min'].append(min_mean_region['Score'])
    result['MaxRegion'].append(max_mean_region['RegName'])
    result['MinRegion'].append(min_mean_region['RegName'])

result_df = pd.DataFrame(result)
plt.rcParams.update({'font.size': 10}) 
fig, ax = plt.subplots(figsize=(12, 6))
result_df.plot(kind='bar', x='Subject', y=['Max', 'Min'], color=['green', 'red'], edgecolor='black', ax=ax)
ax.set_xticks(range(len(subjects)))
ax.set_xticklabels(names, rotation=0)
for i, subj in enumerate(subjects):
    max_val = result_df.at[i, 'Max']
    min_val = result_df.at[i, 'Min']
    max_region = result_df.at[i, 'MaxRegion']
    min_region = result_df.at[i, 'MinRegion']

    ax.text(i, max_val + 12, f'Max: {max_val:.2f} ({max_region})', ha='center')
    ax.text(i, max_val + 8 , f'Min: {min_val:.2f} ({min_region})', ha='center')

ax.set_ylim(100, 200)
ax.set_yticks(range(100, 201, 5))
plt.title('Найвищі та найнижчі середні бали по областях')
plt.xlabel('Предмет')
plt.ylabel('Середній результат')
plt.tight_layout()
plt.show()
```

```python
"""
Show the plot of maximum and minimum average results in 4 subjects
depending on students status.
"""
names = ['Українська мова і література', 'Математика', 'Англійська мова', 'Історія України']
subjects = ['UMLBall100', 'MathBall100', 'EngBall100', 'HistBall100']
result = {'Subject': [], 'Max': [], 'Min': [], 'MaxStatus': [], 'MinStatus': []}
status_dict = {'Студент закладу вищої освіти': 'Студент університу',
'Випускник загальноосвітнього навчального закладу 2021 року': 'Випускник\n2021 року',\
'Випускник минулих років': 'Випускник\nминулих років',\
'Учень (слухач) закладу професійної (професійно-технічної) освіти': 'Здобувач\nпрофесійної\nосвіти',\
'Випускник, який здобуде в 2021 році повну загальну середню освіту в навчальному закладі іншої держави': 'Іноземні заклади'}

for subj in subjects:
    numeric_series = pd.to_numeric(dataset[subj].str.replace(',', '.', regex=False), errors='coerce')
    numeric_df = pd.DataFrame({'Score': numeric_series, 'RegTypeName': dataset['RegTypeName']})
    numeric_df = numeric_df[(numeric_df['Score'].notna()) & (numeric_df['Score'] != 0)]
    result_df = numeric_df.groupby('RegTypeName')['Score'].mean().reset_index()
    max_mean_status = result_df.loc[result_df['Score'].idxmax()]
    min_mean_status = result_df.loc[result_df['Score'].idxmin()]
    result['Subject'].append(subj)
    result['Max'].append(max_mean_status['Score'])
    result['Min'].append(min_mean_status['Score'])
    result['MaxStatus'].append(max_mean_status['RegTypeName'])
    result['MinStatus'].append(min_mean_status['RegTypeName'])

result_df = pd.DataFrame(result)
plt.rcParams.update({'font.size': 14}) 
fig, ax = plt.subplots(figsize=(18, 8))
result_df.plot(kind='line', x='Subject', y=['Max', 'Min'], color=['purple', 'maroon'], marker='o', linestyle=':', ax=ax)

ax.set_xticks(range(len(subjects)))
ax.set_xticklabels(names, rotation=0)
for i, subj in enumerate(subjects):
    max_val = result_df.at[i, 'Max']
    min_val = result_df.at[i, 'Min']
    max_status = result_df.at[i, 'MaxStatus']
    min_status = result_df.at[i, 'MinStatus']

    ax.text(i, max_val + 7, f'Max: {max_val:.2f}\n({status_dict[max_status]})', ha='center')
    ax.text(i, min_val + 7 , f'Min: {min_val:.2f}\n({status_dict[min_status]})', ha='center')

ax.set_ylim(100, 200)
ax.set_yticks(range(100, 201, 5))
plt.title('Середній бал за статусом')
plt.xlabel('Предмет')
plt.ylabel('Середній результат')
plt.tight_layout()
plt.show()
```
